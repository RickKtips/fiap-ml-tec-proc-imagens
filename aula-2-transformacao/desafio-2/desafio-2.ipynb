## Desafio 2


Dado a imagem abaixo (no repositório “imagens/gize.png”), realize a extração da pirâmide, por meio de uma máscara poligonal, para uma outra imagem com fundo na cor verde. 

![alt text](imagens/gize.png)

```python
import cv2
print(cv2.__version__)

import numpy as np

from matplotlib import pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline

plt.style.use('seaborn')
```

```python
image = cv2.imread("imagens/gize.png")

cv2.imshow("Piramide", image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image_plt = mpimg.imread("imagens/gize.png")
plt.imshow(image_plt)
plt.show()
```

```python
mask = image.copy()
image_color = image.copy()

height = image.shape[0]
width = image.shape[1]

mask = np.zeros((height, width), dtype=np.uint8)
pts = np.array([[60,410], [300,150], [520,420]], np.int32)
cv2.fillPoly(mask, [pts], (100, 50, 200))

cv2.imshow("Piramide Mascara", mask)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image_white = np.ones([height, width, 3])
cv2.imshow("Imagem Vazia", image_white)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
res = cv2.bitwise_and(image_color,image_color,mask=mask)
cv2.imshow("Piramide Extraida", res)
cv2.waitKey()
cv2.destroyAllWindows()
```

Substituindo o valor preto (zerado) para outra cor, neste caso verde.

```python
res[np.where((res == [0,0,0]).all(axis = 2))] = [0,255,0]
cv2.imshow("Piramide Extraidass", res)
cv2.waitKey()
cv2.destroyAllWindows()
```
